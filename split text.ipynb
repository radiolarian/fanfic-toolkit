```python
folder = '/Users/noon/Desktop/new media 190/ao3/AO3Scraper/data/attack on titan/fanfics_text_files'
os.chdir(folder)
print os.getcwd()
total = 0
for filename in os.listdir(os.getcwd()):
    if '.txt' in filename:
        print filename
        with open (filename, 'rb') as f_in:
            text = f_in.read()
            paragraphs = re.split('\s{2,}', text)
            i = 0
            for p in paragraphs:
                f = open('paragraphs/'+filename+'-'+str(i)+'.txt','wb')
                f.write(p) 
                f.close()
                i += 1
                total += 1
print 'wrote ', total , ' paragraphs'
```
